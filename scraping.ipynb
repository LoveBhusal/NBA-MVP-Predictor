```python
import requests
import os
import shutil
import time
import random
from bs4 import BeautifulSoup
```

```python
years = list(range(2003,2024))
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
for year in years:
    url = url_start.format(year)
    data = requests.get(url)
    if data.status_code == 200:
        # Save the data to a file
        with open("mvp/{}.html".format(year), "w+", encoding="utf-8") as f:
            f.write(data.text)
    else:
        print(f"Request for year {year} failed with status code {data.status_code}")
    sleep_duration = random.uniform(10, 15)
    time.sleep(sleep_duration)
```

```python
dfs = []
for year in years:
    with open("mvp/{}.html".format(year), "r", encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="mvp")
    html_content = str(mvp_table)
    html_io = StringIO(html_content)
    
    # Use pd.read_html() with the StringIO object
    mvp = pd.read_html(html_io)[0]
    mvp["Year"] = year

    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    url = team_stats_url.format(year)
    data = requests.get(url)

    with open("team/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(data.text)
        print(f"Saved data for {year}")
    
    time.sleep(10)
```

```python
dfs = []

for year in years:
    with open("team/{}.html".format(year), "r", encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    
    # Process Eastern Conference data
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_html_contents = str(e_table)
    e_html_io = StringIO(e_html_contents)
    e_df = pd.read_html(e_html_io)[0]
    e_df["Year"] = year
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    dfs.append(e_df)
    
    # Process Western Conference data
    w_table = soup.find_all(id="divs_standings_W")[0]
    w_html_contents = str(w_table)
    w_html_io = StringIO(w_html_contents)
    w_df = pd.read_html(w_html_io)[0]
    w_df["Year"] = year
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    dfs.append(w_df)
```

```python
teams = pd.concat(dfs)
```

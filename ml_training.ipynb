```python
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("Predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index,row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found / seen)
        seen += 1

    return sum(ps) / len(ps)
```

```python
ap = find_ap(combination)
```

```python
def add_ranks(combination):
    combination = combination.sort_values("Share", ascending = False)
    combination["Rk"] = list(range(1,combination.shape[0]+1))
    combination = combination.sort_values("Predictions", ascending = False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = combination["Rk"] - combination ["Predicted_Rk"]
    return combination
    
```

```python
years = list(range(2003, 2024))

import os
output_folder = "mvp_predictions"
os.makedirs(output_folder, exist_ok=True)

# Create an empty DataFrame to store top predictions
top_predictions_rf = pd.DataFrame(columns=["Year", "Player", "MVP_Prediction_RF"])

# Iterate through each year
for year in years:
    
    # Filter the data for the current year
    players_filtered = stats[stats["Year"] == year][predictors]

    # Make predictions using the Random Forest model
    mvp_predictions = rf.predict(players_filtered)

    # Create a DataFrame with predictions for the current year
    predictions_df_rf = pd.DataFrame({
        "Year": [year] * len(players_filtered),  # Add the current year
        "Player": stats[stats["Year"] == year]["Player"],
        "MVP_Prediction_RF": mvp_predictions,

    })

    # Sort the predictions by MVP_Prediction_RF and reset the index
    predictions_df_rf = predictions_df_rf.sort_values(by="MVP_Prediction_RF", ascending=False)
    predictions_df_rf = predictions_df_rf.reset_index(drop=True)
    predictions_df_rf.index += 1

    # Store only the top 10 predictions
    top_10_predictions = predictions_df_rf.head(10)

    # Remove the "Year" column from the DataFrame
    top_10_predictions = top_10_predictions.drop("Year", axis=1)

    # Concatenate the top predictions for the current year with the result DataFrame
    top_predictions_rf = pd.concat([top_predictions_rf, top_10_predictions])

    file_path = os.path.join(output_folder, f"{year}_predictions.csv")

    # Save the top 10 predictions to a CSV file for the current year
    top_10_predictions.to_csv(file_path, index=False)


top_10_predictions
```
